````python
imp_features=['od280/od315_of_diluted_wines','proline','color_intensity','flavanoids','alcohol']
````

````python
# code here
# Load dataset
import pandas as pd
from sklearn.datasets import load_wine

# load the wine dataset
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target
````

````python
df.describe()
````

````python
df.info()
````

````python
df.isna().sum()
````

````python
#
from matplotlib import pyplot as plt
import seaborn as sns
````

````python
len(data.feature_names)
````

````python
fig,ax=plt.subplots(nrows=len(data.feature_names)//2 +1 , ncols=2,figsize=(20, 15))
ax = ax.flatten()
for i,feature in enumerate(data.feature_names):
  sns.kdeplot(df,x=feature,hue='target',ax=ax[i])
  ax[i].set_title(f'{feature} KDE')
  ax[i].set_xlabel(feature)
  # ax[i].set_ylabel('Density')

plt.subplots_adjust(wspace=0.4, hspace=1)
plt.show()
````

````python
import itertools
two_features=[]
for pair in itertools.combinations(imp_features,2):
  two_features.append(pair)
````

````python
fig,ax=plt.subplots(nrows=len(two_features), ncols=1,figsize=(20,100))
ax = ax.flatten()
for i,feature in enumerate(two_features):
  sns.kdeplot(df,x=feature[0],y=feature[1],hue='target',ax=ax[i])
  ax[i].set_title(f'{feature} KDE')
  ax[i].set_xlabel(feature[0])
  ax[i].set_ylabel(feature[1])

plt.subplots_adjust(wspace=0.4, hspace=1)
plt.show()
````

````python
fig,ax=plt.subplots(nrows=len(two_features), ncols=1,figsize=(20,100))
ax = ax.flatten()
for i,feature in enumerate(two_features):
  sns.jointplot(df,x=feature[0],y=feature[1],hue='target',ax=ax[i])
  ax[i].set_title(f'{feature} KDE')
  ax[i].set_xlabel(feature[0])
  ax[i].set_ylabel(feature[1])

plt.subplots_adjust(wspace=0.4, hspace=1)
plt.show()
````
